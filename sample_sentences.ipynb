```python
sampled_LAM_foundational_df = LAM_foundational_df.sample(17, random_state=42)
```

```python
sampled_LAM_foundational_df
```

```python
sampled_LAM_foundational_sentences_indices = []
for i, row in sampled_LAM_foundational_df.iterrows():
    directory = row['Directory']
    document = row['Document']
    annotations_file_path = f"BAREC-10M/Data/Morphology_and_Readability/{directory}/{document}.json"

    with open(annotations_file_path, 'r', encoding='utf-8') as f:
        annotations = json.load(f)
        annotations = pd.DataFrame(annotations)
    

    sampled_index = annotations[annotations['sents_word_count'] > 5].sample(1, random_state=42).index.item()
    sampled_LAM_foundational_sentences_indices.append(sampled_index)
print(sampled_LAM_foundational_sentences_indices)
```

```python
sampled_LAM_foundational_df['sampled_sentence_index'] = sampled_LAM_foundational_sentences_indices
```

```python
sampled_LAM_specialized_sentences_indices = []
for i, row in sampled_LAM_specialized_df.iterrows():
    directory = row['Directory']
    document = row['Document']
    annotations_file_path = f"BAREC-10M/Data/Morphology_and_Readability/{directory}/{document}.json"

    with open(annotations_file_path, 'r', encoding='utf-8') as f:
        annotations = json.load(f)
        annotations = pd.DataFrame(annotations)
    

    sampled_index = annotations[annotations['sents_word_count'] > 5].sample(1, random_state=42).index.item()
    sampled_LAM_specialized_sentences_indices.append(sampled_index)
print(sampled_LAM_specialized_sentences_indices)
```

```python
sampled_LAM_specialized_df['sampled_sentence_index'] = sampled_LAM_specialized_sentences_indices
```

```python
A_df = pd.concat([sampled_LAM_foundational_df, sampled_LAM_advanced_df, sampled_LAM_specialized_df], ignore_index=True)
A_df
```

```python
A_df['selected_sent_raw'] = None
A_df['selected_sent_word_count'] = None
A_df['selected_sent_words'] = None

for i, row in A_df.iterrows():
    directory = row['Directory']
    document = row['Document']

    annotations_file_path = (
        f"BAREC-10M/Data/Morphology_and_Readability/"
        f"{directory}/{document}.json"
    )

    with open(annotations_file_path, 'r', encoding='utf-8') as f:
        annotations = json.load(f)

    annotations = pd.DataFrame(annotations)

    selected_sentence_index = int(row['sampled_sentence_index'])

    selected_row = annotations.iloc[selected_sentence_index]

    raw_sentence = selected_row['raw_sents']
    word_count = selected_row['sents_word_count']
    words = selected_row['word']   # list of strings

    A_df.at[i, 'selected_sent_raw'] = raw_sentence
    A_df.at[i, 'selected_sent_word_count'] = word_count
    A_df.at[i, 'selected_sent_words'] = words
```

```python
A_df = A_df[[
    'Directory',
    'Document',
    'sampled_sentence_index',
    'Source',
    'Book',
    'Author',
    'Domain',
    'Readership Group',
    'Text Category',
    'average sentence length',
    'selected_sent_raw',
    'selected_sent_word_count',
    'selected_sent_words'
]]
```

```python
A_df.rename(columns={
    'average sentence length': 'average_sentence_length_per_doc',
}, inplace=True)
```
